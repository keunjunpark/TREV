```python
import gurobipy as gp
from gurobipy import GRB
import networkx as nx

def maxcut_value(G) -> float:
    """
    Solve weighted Max-Cut on an undirected NetworkX graph using Gurobi.

    Args:
        G: networkx.Graph
           Undirected graph. Edge weights read from attribute 'weight' (default 1.0).

    Returns:
        float: Optimal Max-Cut value.
    """
    # if not isinstance(G, nx.Graph) or G.is_directed():
    #     raise ValueError("G must be an undirected networkx.Graph")

    model = gp.Model("MaxCut")
    model.Params.OutputFlag = 0  # silence solver output

    # Binary variable x[i] ∈ {0,1} indicates partition side of node i
    #nodes = list(G.nodes())
    # Number of nodes
    nodes = set()
    for e,w in G:
        u,v = e
        nodes.add(u)
        nodes.add(v)
    n = len(nodes)

    x = model.addVars(nodes, vtype=GRB.BINARY, name="x")
    # Objective: Maximize the cut value
    model.setObjective(
        gp.quicksum(w * (x[u] + x[v] - 2 * x[u] * x[v]) for (u, v), w in G),
        GRB.MAXIMIZE
    )

    # Optimize the model
    model.optimize()


    if model.Status != GRB.OPTIMAL:
        raise RuntimeError(f"Gurobi did not find an optimal solution (status={model.Status}).")

    return float(model.ObjVal)
```

```python
def _first_opt_idx_and_plateau_len(cuts: np.ndarray, opt: float, tol: float = 0.0):
    """Return (first_index, plateau_len) where cuts stay >= opt - tol from that index onward.
       If no hit, return (None, 0)."""
    hit = np.where(cuts >= opt - tol)[0]
    if hit.size == 0:
        return None, 0
    i0 = int(hit[0])
    # plateau = consecutive points from i0 while condition holds
    mask = (cuts >= opt - tol)
    j = i0
    n = cuts.shape[0]
    while j < n and mask[j]:
        j += 1
    return i0, (j - i0)

def _best_series_to_cuts(best_series, N: int, seed: int, all_graph):
    arr = np.asarray(best_series)
    if all_graph is None or np.issubdtype(arr.dtype, np.number):
        return arr.astype(float)
    graph_lookup = {(g['N'], g['graph_id']): g['G'] for g in all_graph}
    G = graph_lookup[(N, seed)]
    temp_tsp = Tsp(G)
    val = solve_tsp_gurobi(nx.to_numpy_array(temp_tsp.graph))
    qp = temp_tsp.to_quadratic_program()
    qp2qubo = QuadraticProgramToQubo()
    qubo_temp = qp2qubo.convert(qp)
    maximum_cut = val[1]
    cuts = []
    recent_best = 0
    for s in arr:
        #s = s[::-1]
        z = temp_tsp.interpret(np.array(list(s), dtype=int))
        # print(z)
        # print(s)
        is_pos = True
        for elm in z:
            if type(elm) == list:
                is_pos = False
                break
        if not is_pos:
            #cuts.append(recent_best/maximum_cut)
            cuts.append(recent_best)
        else:
            adj_matrix = nx.to_numpy_array(temp_tsp.graph)
            val = temp_tsp.tsp_value(z, adj_matrix)
            cuts.append(val)
            recent_best = val
    return  np.array(cuts)

def _final_avg_plateau_aware(
    run_id: str,
    outdir: str,
    *,
    N: int,
    seed: int,
    gurobi_opt: float,
    all_graph,
    tol: float = 0.0,
    min_plateau: int = 10,     # “stays optimal ~10 iterations”
    plateau_tail: int | None = None,  # if set, average only the last K of the plateau
    window_center_half: int = 5,      # used if hit but short plateau
    window_tail: int = 10,            # used if never hits optimum
) -> float:
    d = read_iter_csv(run_id, outdir=outdir)
    best_col = "best_result" if "best_result" in d else ("best_results" if "best_results" in d else None)
    if best_col is None:
        raise KeyError("Neither 'best_result' nor 'best_results' found in iter CSV.")

    cuts = _best_series_to_cuts(d[best_col], N=N, seed=seed, all_graph=all_graph)
    # 1) Plateau-aware path
    i0, plen = _first_opt_idx_and_plateau_len(cuts, gurobi_opt, tol=tol)
    if i0 is not None:
        if plen >= min_plateau:
            if plateau_tail is not None and plateau_tail > 0:
                start = i0 + max(0, plen - plateau_tail)
                end   = i0 + plen
                return float(np.mean(cuts[start:end]))
            # average entire plateau
            return float(np.mean(cuts[i0:i0+plen]))
        # 2) Hit but short plateau → centered window around first hit
        lo = max(0, i0 - window_center_half)
        hi = min(len(cuts), i0 + window_center_half + 1)
        return float(np.mean(cuts[lo:hi]))

    # 3) Never hit optimum → tail average
    k = min(window_tail, len(cuts))
    return float(np.mean(cuts[-k:]))
```

```python
import numpy as np
import gurobipy as gp
from gurobipy import GRB

def solve_tsp_gurobi(distance_matrix):
    """
    Solve TSP given distance matrix using Gurobi.
    """
    N = distance_matrix.shape[0]
    model = gp.Model("TSP")

    # Variables: x[i,j] = 1 if traveling from i to j
    x = model.addVars(N, N, vtype=GRB.BINARY, name="x")

    # MTZ variables for subtour elimination
    u = model.addVars(N, vtype=GRB.CONTINUOUS, lb=0, ub=N-1, name="u")

    # Objective: minimize total distance
    model.setObjective(
        gp.quicksum(distance_matrix[i, j] * x[i, j] for i in range(N) for j in range(N)),
        GRB.MINIMIZE
    )

    # Constraints: Exactly one outgoing and incoming edge
    for i in range(N):
        model.addConstr(gp.quicksum(x[i, j] for j in range(N) if i != j) == 1)
    for j in range(N):
        model.addConstr(gp.quicksum(x[i, j] for i in range(N) if i != j) == 1)

    # MTZ subtour elimination
    for i in range(1, N):
        for j in range(1, N):
            if i != j:
                model.addConstr(u[i] - u[j] + (N-1)*x[i, j] <= N-2)

    model.setParam('OutputFlag', 0)  # Silent mode
    model.optimize()

    if model.status == GRB.OPTIMAL:
        sol = model.getAttr('x', x)
        # Recover tour
        tour = []
        current_city = 0
        visited = set()
        while len(visited) < N:
            visited.add(current_city)
            for j in range(N):
                if j != current_city and sol[current_city, j] > 0.5:
                    tour.append(current_city)
                    current_city = j
                    break
        tour.append(current_city)
        cost = model.objVal
        return tour, cost
    else:
        raise Exception("No optimal solution found.")
#
# # --- Example Usage ---
# if __name__ == "__main__":
#     N =2     # number of cities
#     dist = generate_random_tsp(N, seed=42)
#     print("Randomly generated TSP distance matrix:\n", dist)
#
#     tour, cost = solve_tsp_gurobi(dist)
#     print("\nOptimal tour:", tour)
#     print("Optimal cost:", cost)




```

```python
import numpy as np
import pandas as pd
import networkx as nx
from scipy.optimize import linear_sum_assignment

# --------- Helpers: decode + length + score ---------------------------------

def bits_to_assignment_matrix(bitstr: str, n: int) -> np.ndarray:
    """
    QUBO TSP layout assumed as n x n (rows=cities, cols=positions).
    bitstr is length n*n; element k maps to (i, t) where k = i*n + t  (or your convention).
    """
    # If your packing is different, adjust the reshape order.
    x = np.array(list(bitstr), dtype=int)
    if x.size != n * n:
        raise ValueError(f"bitstring length {x.size} != n*n ({n*n})")
    X = x.reshape(n, n)  # rows=cities, cols=positions
    return X

def decode_feasible_tour_from_matrix(P: np.ndarray) -> list[int]:
    """
    P can be one-hot (from bits) or soft (marginals). Hungarian gives a permutation.
    Returns a list of city indices in visiting order (positions 0..n-1).
    """
    # Hungarian minimizes cost; maximize P by minimizing -P
    row_ind, col_ind = linear_sum_assignment(-P)
    # col_ind is assigned position per row (city). Sort by position to get the order.
    order = np.argsort(col_ind)          # positions 0..n-1
    tour = row_ind[order].tolist()       # city at each position
    return tour

def tour_length(tour: list[int], adj: np.ndarray) -> float:
    n = len(tour)
    return float(sum(adj[tour[k], tour[(k+1) % n]] for k in range(n)))

# --------- Your aggregate_over_seeds, with the scoring block replaced -------

def aggregate_over_seeds(run_ids, g, cut_off_it, denoise="running_best",
                         rolling_window=10, outdir="bench_out", all_graph=None):
    """
    Return iteration array, median curve and lower/upper band across seeds.
    Curves are Feasible-Tour Score (FTS) = L_star / L_t (higher is better; 1.0 = optimal).
    """
    curves = []
    iters_ref = None
    graph_lookup = {(g['N'], g['graph_id']): g['G'] for g in all_graph}

    for rid in run_ids:
        df = pd.read_csv(f"{outdir}/summary.csv")
        summary = df[df["run_id"] == rid].iloc[0]
        seed = summary["seed"]
        N = summary["N"]
        if all_graph is not None:
            g = graph_lookup[(N, seed)]

        d = read_iter_csv(rid, outdir=outdir)
        best_col = "best_result" if "best_result" in d else "best_results"
        iters = np.array(d["iter"][:cut_off_it])

        # --- Build TSP instance and get L_star (Gurobi optimal tour length) ---
        temp_tsp = Tsp(g)  # g is your NetworkX graph (as you used before)
        adj_matrix = nx.to_numpy_array(temp_tsp.graph)

        # If you already have a solver returning (tour, length):
        L_star = solve_tsp_gurobi(adj_matrix)[1]

        # --- NEW: compute Feasible-Tour Score at every iteration --------------
        fts = []
        for s in d[best_col][:cut_off_it]:
            # keep your original reversal if needed

            # Make an assignment matrix from bits (or probabilities if you store them)
            X = bits_to_assignment_matrix(s, n=adj_matrix.shape[0])

            # Always decode a valid tour with Hungarian
            tour = decode_feasible_tour_from_matrix(X)
            print(tour)
            # Evaluate its length and turn into a normalized, increasing score
            L_t = tour_length(tour, adj_matrix)
            score = (L_star / L_t) if L_t > 0 else 0.0
            fts.append(score)

        # Denoise (optional)
        if denoise == "running_best":
            fts = np.maximum.accumulate(fts)
        elif denoise == "rolling":
            fts = pd.Series(fts).rolling(rolling_window, min_periods=1).mean().to_numpy()

        curves.append(np.array(fts))
        if iters_ref is None:
            iters_ref = iters

    # pad and aggregate
    max_len = max(len(c) for c in curves)
    curves = [np.pad(c, (0, max_len - len(c)), constant_values=c[-1]) for c in curves]
    curves = np.vstack(curves)
    median = np.median(curves, axis=0)
    q25 = np.percentile(curves, 25, axis=0)
    q75 = np.percentile(curves, 75, axis=0)
    return iters_ref, median, q25, q75
```
